```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,10,20)
x
```

```python
y=randint(1,50,20)
y
```

```python
plt.plot(x,y,color='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Plot')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'ro',markersize=10)
plt.subplot(1,2,2)
plt.plot(x,y,'b*-')
```

**matlab vs matplotlib**

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Set Title of Random')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y,'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Y plot')
ax2.plot(x,y,'g')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('X plot')
```

```python
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y,'b')
ax[1].plot(x,y,'r')
```

```python
fig,ax=plt.subplots(1,2)
col=['r','g']

for i,axes in enumerate(ax):
    axes.plot(x,y,col[i])
fig.tight_layout()
```

```python
fig,ax=plt.subplots(figsize=(8,4),dpi=100)
ax.plot(x,y,'r',label='y*x')
ax.plot(x,y,'b',label='y*x' )
ax.set_label('X')
ax.set_label('Y')
ax.set_title('Random no Plot')
ax.legend(ioc=5)
fig.savefig('random.png',dpi=300)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,'b.--',markersize=12,linewidth=3)
```
